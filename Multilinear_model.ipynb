```python
import pandas as pd
import numpy as np

# from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
# from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_excel("/content/SalesData.xlsx")
```

```python
data.head()
```

```python
data.info()
```

```python
dummie_gender = pd.get_dummies(data.Gender, drop_first=True).astype(int)
dummie_Region = pd.get_dummies(data.Region, drop_first=True).astype(int)
```

```python
data = data.drop(["Gender", "Region"], axis=1)
```

```python
data
```

```python
data = pd.concat([data,dummie_gender, dummie_Region], axis=1)
```

```python
data
```

```python
x = data.drop("Sales", axis=1)
y = data.Sales
```

```python
xtrain,xtest, ytrain, ytest = train_test_split(x,y, random_state=20, test_size=0.2)
```

```python
xtrain = sm.add_constant(xtrain)
xtest = sm.add_constant(xtest)
```

```python
xtrain.head()
```

```python
mod = sm.OLS(ytrain, xtrain).fit()
```
