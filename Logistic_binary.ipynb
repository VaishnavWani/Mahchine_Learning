```python
import pandas as pd
import numpy as np

# from sklearn.model_selection import train_test_split
# import statsmodels.api as sm

from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
dt = pd.read_excel("/content/drive/MyDrive/CDAC_DataBook.xlsx", sheet_name="mtcars")
```

```python
dt.head(1)
```

```python
dt = dt.drop(["cyl",'disp','drat','qsec','vs','carb'], axis = 1)
```

```python
dt.head()
```

```python
am_dummies = pd.get_dummies(dt.am, drop_first = True,  prefix="am_").astype(int)
gear_dummies = pd.get_dummies(dt.gear, drop_first = True,  prefix="gear_").astype(int)
```

```python
dt = pd.concat([dt,am_dummies,gear_dummies],axis = 1)
```

```python
dt.head()
```

```python
dt = dt.drop(["am","gear"], axis = 1)
```

```python
dt.head()
```

```python
dt['mpg'].median()
```

```python
dt['resp_mpg'] = np.where(dt['mpg']>dt['mpg'].median(),1,0)
```

```python
dt.head()
```

```python
dt = dt.drop(['mpg'],axis=1)
```

```python
dt.head()
```

```python
x = dt.drop(['resp_mpg'], axis = 1)
y = dt.resp_mpg
```

```python
y.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, random_state = 20, test_size=0.2 )
```

```python
xtrain  = sm.add_constant(xtrain)
xtest = sm.add_constant(xtest)
```

```python
xtrain.head()
```

```python
model = sm.Logit(ytrain , xtrain).fit()
```

```python
print(model.summary())
```
